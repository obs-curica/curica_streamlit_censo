```python
import pandas as pd
import os

# 1. Leitura correta do CSV, com separador ponto-e-vírgula
#df_fundeb_ac = pd.read_csv('data/fundeb/df_fundeb_original.csv', sep=';', encoding='utf-8', low_memory=False)

# 2. Corrigir as colunas de valores numéricos
colunas_valores = ['2019', '2020', '2021', '2022', '2023', '2024']

for coluna in colunas_valores:
    df_fundeb_ac[coluna] = (
        df_fundeb_ac[coluna]
        .astype(str)
        .str.replace('.', '', regex=False)     # remove separador de milhar
        .str.replace(',', '.', regex=False)     # troca vírgula decimal por ponto
        .str.strip()                             # remove espaços extras
        .astype(float)
    )

# 3. Agrupar por Município
df_fundeb_ac = (
    df_fundeb_ac
    .groupby('Município')[colunas_valores]
    .sum()
    .reset_index()
)

# 4. Salvar CSV com valores convertidos
# os.makedirs('data/fundeb', exist_ok=True)
# df_fundeb_ac.to_csv('data/fundeb/df_fundeb_ac.csv', sep=';', index=False, encoding='utf-8', float_format="%.2f")

# print("CSV salvo com sucesso.")

df_fundeb_ac.head(5)
```

```python
# Convertendo df_fundeb_ac para formato longo
df_fundeb_ac_longo = df_fundeb_ac.melt(
    id_vars=['Município'],        # Mantém Município fixo
    value_vars=['2019', '2020', '2021', '2022', '2023', '2024'],  # Colunas que queremos "derreter"
    var_name='Ano',               # Nome da nova coluna para os anos
    value_name='Total_Repassado'  # Nome da nova coluna para os valores somados
)

# Exibir o resultado
df_fundeb_ac_longo.head(5)

```

```python
# 4. Salvar CSV com valores convertidos
os.makedirs('data/fundeb', exist_ok=True)
df_fundeb_ac.to_csv('data/fundeb/df_fundeb_ac.csv', sep=';', index=False, encoding='utf-8', float_format="%.2f")

print("CSV salvo com sucesso.")
```
